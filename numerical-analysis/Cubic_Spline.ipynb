```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt
```

```python
# Display grid of 16 circles
def DrawGrid():
    plt.figure(figsize=(6,6))
    [gx, gy] = np.meshgrid([0, 20, 40, 60], [0, 20, 40, 60])
    plt.plot(gx, gy,'o', color='lightgray', markersize=10); plt.axis('square');
```

```python
DrawGrid()
```

## (a) Fit Points with a Spline

```python
# === YOUR CODE HERE ===
time = [0, 795, 1447, 1881, 2172, 2897]
x = [20,34,57,47,7,22]
y = [18,54,53,30,35,61]
spl1 = make_interp_spline(time, x, bc_type="clamped")
spl2 = make_interp_spline(time, y, bc_type="clamped")
t = [t for t in range(0,2898,2)]
xx = spl1(t)
yy = spl2(t)
```

## (b) Plot the Spline

```python
# === YOUR CODE HERE ===
DrawGrid()
plt.plot(x,y,'o')
plt.plot(xx, yy)
```
